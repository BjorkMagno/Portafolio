## Importación de paquetería.

```python
import pandas as pd
```

## 1. Importación de documentos.

```python
archivo = '/content/datos_hosting.json'
datos_json = pd.read_json(archivo)
```

```python
datos_json.head()
```

Notemos que es un archivo anidado por lo que vamos a normalizarlo.

```python
# Normalización de la lectura anterior.
datos_json = pd.json_normalize(datos_json['info_inmuebles'])
datos_json.head()
```

```python
# Comprobación de ser data frame.
datos_json.info()
```

## 2. Datos numéricos.

```python
columnas = list(datos_json.columns)
columnas
```

```python
datos_explotado = datos_json.explode(columnas[3:])
datos_explotado
```

```python
datos_explotado.reset_index(inplace = True, drop=True)
datos_explotado.head()
```

```python
datos_explotado.info()
```

```python
# Cambiar tipo de dato al correcto.
import numpy as np
```

```python
datos_explotado['max_hospedes'] = datos_explotado['max_hospedes'].astype(np.int64)
```

```python
datos_explotado.info()
```

```python
col_numericas = ['cantidad_baños', 'cantidad_cuartos','cantidad_camas']
```

```python
datos_explotado[col_numericas] = datos_explotado[col_numericas].astype(np.int64)
```

```python
datos_explotado.info()
```

```python
datos_explotado['evaluacion_general'] = datos_explotado['evaluacion_general'].astype(np.float64)
```

```python
datos_explotado.info()
```

```python
datos_explotado['precio']
```

```python
datos_explotado['precio'] = datos_explotado['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
datos_explotado.head()
```

```python
datos_explotado['precio'] = datos_explotado['precio'].astype(np.float64)
```

```python
datos_explotado.info()
```

```python
datos_explotado[['cuota_deposito', 'cuota_limpieza']] = datos_explotado[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$','').replace(',','').strip()).astype(np.float64)
```

```python
datos_explotado.info()
```

## Datos de texto.

```python
datos_explotado.head()
```

```python
datos_explotado['descripcion_local']=datos_explotado['descripcion_local'].str.lower()
```

```python
datos_explotado.head()
```

```python
datos_explotado['descripcion_local'][3169]
```

```python
datos_explotado['descripcion_local']=datos_explotado['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
datos_explotado['descripcion_local']=datos_explotado['descripcion_local'].str.replace('(?<!w)-(?!\w)',' ',regex=True)
```

```python
datos_explotado['descripcion_local'][3169]
```

```python
datos_explotado['descripcion_local']=datos_explotado['descripcion_local'].str.split()
```

```python
datos_explotado['comodidades']=datos_explotado['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
datos_explotado['comodidades']=datos_explotado['comodidades'].str.split(',')
datos_explotado
```

```python
datos_explotado['descripcion_vecindad'][3814]
```

```python
datos_explotado['descripcion_vecindad']=datos_explotado['descripcion_vecindad'].str.lower()
datos_explotado['descripcion_vecindad']=datos_explotado['descripcion_vecindad'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
datos_explotado['descripcion_vecindad']=datos_explotado['descripcion_vecindad'].str.replace('(?<!w)-(?!\w)','',regex=True)
datos_explotado['descripcion_vecindad']=datos_explotado['descripcion_vecindad'].str.split()
```
