### Análisis exploratorio de datos.

El proceso de **Análisis Exploratorio de Datos (EDA**) consiste en comprender cómo están estructurados los datos que queremos analizar.

Es un proceso de naturaleza **investigativa**, donde intentamos entender diversas características, como: los valores presentes en las columnas, los tipos de estructura de datos, verificar si son datos cualitativos o cuantitativos, si hay valores faltantes o inusuales.

#### ¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

````python
# Primero realizamos una visualización estádar.
datos.head()
````

Requerimos obtener el promedio de los datos en la columna **'Valor'** por cada tipo existente en la columna **'Tipo'**, naturalmente pensamos en

```
.mean()
```

sin embargo, este nos da el *promedio general de la columna valor*.

````python
datos['Valor'].mean()
````

Para solucionar el problema realizamos la agrupación en tablas de la información según el tipo en la coluumna **'Tipo'**, por lo que requerimos el uso de

```
.groupby('Columna')
```

La función *agrupa la información de los datos que conciden en la columna señalada*, es decir, si yo tengo dos o  más valores iguales en una columna la información de estos se agrupan.

````python
datos.groupby('Tipo')
````

Regresando al problema deseamos combinar las funciones anteriores, sin embargo, la tabla contiene letras que la función mean no acepta. Para solucionarlo le pedimos que solo tome encuenta los valores númericos.

```
datos.groupby('Tipo').mean(numeric_only=True)
```
No obstante, esto hace el *promedio de la agrupación de todos los valores numericos por columnas*.


````python
datos.groupby('Tipo').mean(numeric_only=True)
````

Finalmenete se soluciona al agrupar los valores y obtener el promedio de la columna correspondiente.

````python
# Promedio de una tabla que agrupa el valor.
datos.groupby('Tipo')['Valor'].mean()
````

````python
# Promedio de una tabla que agrupa el valor de forma ascedente.
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
````

##### Valores promedio de alquiler por tipo de inmueble.

````python
# Almacenamiento de los datos obetenidos y graficación.
df_tipo_precio = datos.groupby('Tipo')['Valor'].mean().sort_values()
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='purple')
````

Sin embargo, necesitamos segregar la información, primero necesitamos saber los tipos inmuebles no comerciales y los imuebles comerciales.

````python
# Obtiene los inmuebles que tenemos sin repetición.
datos.Tipo.unique()
````

````python
# Definimos un arreglo de inmuebles comerciales.
inmuebles_comerciales = ['Edificio Completo','Tienda/Salón', 'Casa Comercial',
                         'Casa de Vecindad','Terreno Estándar', 'Cochera/Estacionamiento',
                          'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial', 'Hotel', 'Loteo/Condominio','Industria']
````

````python
# Devuelve un DataFrame con todos los inmuebles comerciales
datos.query('@inmuebles_comerciales in Tipo')
````

````python
# Devuelve un Data Frame con la iformación de los inmuebles que no estan en el arreglo de inmuebles comerciales.
df_inmuebles_no_comerciales=datos.query('@inmuebles_comerciales not in Tipo')
df_inmuebles_no_comerciales.head()
````

````python
# Almacenamiento de los datos obetenidos y graficación.
df_tipo_precio_inc = df_inmuebles_no_comerciales.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio_inc.plot(kind='barh',figsize=(12,8),color='purple')
````

#### ¿Cuál es el porcentaje de cada tipo de propiedad en nuestra base de datos?

````python
# Cantidad de inmuebles no comerciales.
df_inmuebles_no_comerciales.Tipo.value_counts()
````

````python
# Cantidad de inmuebles no comercialesm en porcentaje.
df_inmuebles_no_comerciales.value_counts(normalize=True)
````

##### Porcentaje de cada tipo de inmueble presente en la base de datos.

````python
# Almacenamiento de los datos obetenidos y graficación.
porcentaje_departamentos = df_inmuebles_no_comerciales.Tipo.value_counts(normalize=True).to_frame()
porcentaje_departamentos.plot(kind='barh',figsize=(12,8),color='purple',xlabel='Tipo de inmueble',ylabel='Porcentaje')
````

Por idicaciónes de la ML requerimos unicamente los inmuebles del tipo departamento, para obtener de un tipo en especifico, realizamos la segregación de el.

````python
# Agrupación de unicamente inmuebles no comerciales del tipo departamento.
df_departamentos = df_inmuebles_no_comerciales.query('Tipo == "Departamento"')
df_departamentos.head()
````

````python
df_departamentos.shape
````

#### Calculo del promedio de habitaciones por departamento.

````python
# Promedio de las habitaciones por departamento.
df_habitaciones = df_departamentos.groupby('Habitaciones')[['Valor']].mean().sort_values('Valor')
df_habitaciones
````
