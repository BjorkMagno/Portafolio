# **ADMINISTRACIÓN DE CONDOMINIOS**

Administrar condominios es una tarea que requiere mucha atención y organización. Entre las diversas responsabilidades de gestión se encuentra el cobro del alquiler a los inquilinos. Para garantizar la buena salud financiera de la empresa, es fundamental que estos pagos se realicen de forma regular y puntual. Sin embargo, sabemos que esto no siempre sucede.

Teniendo esto en cuenta, realicemos un procesamiento de datos con el objetivo de **analizar el retraso en el pago del alquiler en el condominio de algunos residentes**. La base de datos es [dados_locacao_imoveis.json](https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json), que contiene información sobre:
 - el departamento de los inquilinos,
 - el día acordado para el pago del alquiler,
 - el día en que se realiza el pago del alquiler y
 - el monto del alquiler.

## Importación de paquetería

```python
import pandas as pd
import requests
import json
```

```python
import numpy as np
```

## Importación de datos.

```python
# Almacenamiento del archivo.
dados_vendas_clientes = requests.get('https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json')
archivo_json = json.loads(dados_vendas_clientes.text)
# Almacenamiento usual, normalizando.
archivo_json = pd.json_normalize(archivo_json)
archivo_json.head()
```

```python
archivo_normalizado = pd.json_normalize(archivo_json['dados_locacao'][0])
```

```python
archivo_normalizado
```

## Datos numéricos.

```python
columnas = list(archivo_normalizado.columns)
columnas
```

```python
archivo_normalizado = archivo_normalizado.explode(columnas[1:])
archivo_normalizado
```

```python
archivo_normalizado.reset_index(drop=True, inplace=True)
archivo_normalizado
```

```python
archivo_normalizado.info()
```

```python
archivo_normalizado['valor_aluguel']
```

```python
archivo_normalizado['valor_aluguel'] = archivo_normalizado['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace(',', '.').replace(' reais', ''))
archivo_normalizado
```

```python
archivo_normalizado['valor_aluguel'] = archivo_normalizado['valor_aluguel'].astype(np.float64)
archivo_normalizado.info()
```

## Datos numéricos.

Buscando explicar la organización de la identificación de los apartamentos, durante la creación del conjunto de datos se añadió el texto (blocoAP). Este texto informa que los nombres de los apartamentos están organizados con la letra mayúscula seguida del número del apartamento. Sin embargo, esto no aporta ninguna información a nuestros datos, por lo que resulta interesante eliminar este texto del conjunto de datos

```python
archivo_normalizado['apartamento'] = archivo_normalizado['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)
archivo_normalizado
```

## Datos tiempo

```python
archivo_normalizado['datas_de_pagamento'] = pd.to_datetime(archivo_normalizado['datas_de_pagamento'], format='%d/%m/%Y')
archivo_normalizado['datas_combinadas_pagamento'] = pd.to_datetime(archivo_normalizado['datas_combinadas_pagamento'], format='%d/%m/%Y')
archivo_normalizado.head()
```

```python
archivo_normalizado['atraso_pagamento'] = archivo_normalizado['datas_de_pagamento'] - archivo_normalizado['datas_combinadas_pagamento']
archivo_normalizado.head()
```

```python
media_atraso = archivo_normalizado.groupby(['apartamento'])['atraso_pagamento'].mean()
media_atraso
```
